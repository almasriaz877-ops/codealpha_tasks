```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.amazon.com/Apple-Wireless-Bluetooth-Headphones-Cancellation/dp/B0DLY219Q7/ref=sr_1_1?crid=8GS7K5IQAY7F&dib=eyJ2IjoiMSJ9.kfVqMPQ2rgZ39R6D_N9MHp6f7v_6fFevbeBpB-bJgISk_X6EhtyziQ3pbfKSzgTWPzsk9909Rxp4XSHfK69VJNL2w5zT-fGFfNsGHEQjm8ZkFJaCElyFjTnvVTc3GPdRUHk8UiBnC3Axc7Nl33bIR4YIiAZvul8tUo9xI__FSGA4X3DoHc8iReawYNafn6wEitdKP0D10zDabERDZoTzMqHGHEB04m7u3tpymgJsdA8.PKyU9yVF08kqMGiXkduSG01s3rAUdu8TB5aq9M_lRrc&dib_tag=se&keywords=apple+airpods&qid=1782373569&sprefix=appl%2Caps%2C501&sr=8-1"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/149.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9"
}
response = requests.get(url, headers=headers)
if response.status_code == 200:
    print("response.status_code")
else: 
    print("fetching rrror", response.status_code)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
print(soup.title.text)
```

```python
title = soup.find("span", id="productTitle")
if title:
    title = title.get_text(strip=True)
    print(title)
else:
    print("Title Not Fount")
```

```python
price = soup.find("span", class_="a-price-whole")
if price:
    price = price.get_text(strip=True)
    print(price)
else:
    print("price not found")
    
```

```python
rating = soup.find("span", class_="a-icon-alt")
if rating:
    rating = rating.get_text(strip=True)
    print(rating)
else:
    print("rating not found")
```

```python
reviews = soup.find("span", id="acrCustomerReviewText")
if reviews:
    reviews = reviews.get_text(strip=True)
    print(reviews)
else:
    print("reviews not found")
```

```python
availability = soup.find("div", id="availability")
if reviews:
    availability = availability.get_text(strip=True)
    print(availability)
else:
    print("reviews not found")
```

```python
brand = soup.find("a", id="bylineInfo")
if brand:
    brand = brand.get_text(strip=True)
    print(brand)
else:
    print("brand not found")
```

```python
image = soup.find("img", id="landingImage")
if image:
    print(image.get("src"))
else:
    print("image not found")
```

```python
data = {
    "Product Name" : [title],
    "Price" : [price],
    "Rating" : [rating],
    "Reviews" : [reviews],
    "Availability" : [availability],
    "Brand" : [brand],
    "ImageURL" : [image.get("src")]
}
df = pd.DataFrame(data)
df
```

```python
df["Price"] = df["Price"].str.replace(",","")
df["Price"] = df["Price"].astype(float)
df
```

```python
print(df.dtypes)
```

```python
df["Rating"] = df["Rating"].astype(str)
df["Rating"] = df["Rating"].str.extract(r'(\d+\.\d+)')
df["Rating"] = df["Rating"].astype(float)
```

```python
df["Reviews"] = df["Reviews"].str.replace("(", "", regex=False)
df["Reviews"] = df["Reviews"].str.replace(")", "", regex=False)
df["Reviews"] = df["Reviews"].str.replace(",", "", regex=False)
df["Reviews"] = df["Reviews"].astype(int)
```
